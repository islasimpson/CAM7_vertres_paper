### Read in the data

```python
#ystart = 1979 ; yend = 2022
ystart=1850 ; yend=2099
```

```python
pathout="/glade/campaign/cgd/cas/islas/python/L83/DATA_SORT/MJO_OLR/"
```

```python
for imem in np.arange(1,3+1,1):
    basepathhist="/glade/campaign/cesm/development/cvcwg/cvwg/L83/timeseries/b.e21.BHISTcmip6.f09_g17.L83_cam6."+str(imem).zfill(3)+"/atm/proc/tseries/day_1/"
    # zonal mean U
    uzmhist = xr.open_mfdataset(basepathhist+"*.Uzm.*.nc")
    uzmhist = read.fixcesmtime_daily(uzmhist)
    fluthist = xr.open_mfdataset(basepathhist+"*.FLUT.*.nc")
    fluthist = read.fixcesmtime_daily(fluthist)
    fluthist['lon'] = uzmhist.lon ; fluthist['lat'] = uzmhist.lat
    
    basepathssp="/glade/campaign/cesm/development/cvcwg/cvwg/L83/timeseries/b.e21.BSSP370cmip6.f09_g17.L83_cam6."+str(imem).zfill(3)+"/atm/proc/tseries/day_1/"
    uzmssp = xr.open_mfdataset(basepathssp+"*.Uzm.*.nc")
    uzmssp = read.fixcesmtime_daily(uzmssp)
    uzmssp['lon'] = uzmhist.lon ; uzmssp['lat'] = uzmhist.lat
    flutssp = xr.open_mfdataset(basepathssp+"*.FLUT.*.nc")
    flutssp = read.fixcesmtime_daily(flutssp)
    flutssp['lon'] = uzmhist.lon ; flutssp['lat'] = uzmhist.lat

    uzm = xr.concat([uzmhist, uzmssp], dim='time')
    flut = xr.concat([fluthist, flutssp], dim='time')
    
    uzm = uzm.sel(time=slice(str(ystart)+"-01-01",str(yend+1)+"-12-31"))
    uzm = uzm.where( uzm.time.dt.hour == 12, drop=True )
    uzm = uzm.Uzm.isel(zlon=0)

    flut = flut.sel(time=slice(str(ystart)+"-01-01",str(yend+1)+"-12-31"))
    flut = flut.where( flut.time.dt.hour == 12, drop=True)
    flut = flut.FLUT

    alluzm = []
    allflut = []
    for iyear in np.arange(ystart, yend+1, 1):
        print(iyear)
        uzmuse = uzm.sel(time=slice(str(iyear)+"-12-01",str(iyear+1)+"-02-28")).mean('time')
        alluzm.append(uzmuse)

        flutuse = flut.sel(time=slice(str(iyear)+"-11-01",str(iyear+1)+"-03-31"))
        flutclim = flutuse.mean('time')
        flutanoms = flutuse - flutclim
        flutanoms['time'] = pd.date_range("1970-11-01","1971-03-31")
        mjoolr = filt.wkfilter(flutanoms,0.15,1,5,20,100,spd=1)
        mjoolr = mjoolr.sel(time=slice("1970-12-01","1971-02-28"))
        allflut.append(mjoolr)

    year = xr.DataArray(np.arange(ystart+1,yend+2,1), coords=[np.arange(ystart+1,yend+2,1)], dims=['year'], name='year')
    allflut = xr.concat(allflut, dim=year)
    alluzm = xr.concat(alluzm, dim=year)
    allflut = allflut.rename('MJO_OLR')
    alluzm = alluzm.rename('Uzm')

    ds = xr.merge([allflut, alluzm])
    ds.to_netcdf(pathout+'MJO_OLR_'+str(ystart)+'_'+str(yend+1)+'_'+str(imem).zfill(3)+'.nc')
```
