```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

from CASutils import qbo_utils as qbo
from CASutils import plotting_utils as myplots
from CASutils import colorbar_utils as cbars
import sys
```

```python
plotpath="/project/cas/islas/python_plots/L83_paper/FIGURES/SUPP/"
```

```python
basepath="/project/cas/islas/python_savs/L83_paper/DATA_SORT/QBOcomposites/90day_beforeW_powerspec/"
```

```python
era5_50 = xr.open_dataset(basepath+'ERA5_50hpa_5S5N_90days.nc')
jra55_50 = xr.open_dataset(basepath+'JRA55_50hpa_5S5N_90days.nc')
merra2_50 = xr.open_dataset(basepath+'MERRA2_50hpa_5S5N_90days.nc')
```

```python
dcurves = xr.open_dataset("/project/cas/islas/python_savs/L83_paper/DATA_SORT/dispersion_curves/dispersion_curves.nc")
```

```python
def oplot_dcurves(ax, dcurves, component='symmetric'):
    if (component == "symmetric"):
        for i in np.arange(0,dcurves.h_e.size,1):
            ax.plot(dcurves.k, dcurves.w_kelvin.isel(h_e=i), color='gray')
            ax.plot(dcurves.k, dcurves.w_ig_n1.isel(h_e=i), color='gray')
            ax.plot(dcurves.k, dcurves.w_etrap.isel(h_e=i), color='gray')
    else:
        for i in np.arange(0,dcurves.h_e.size,1):
            ax.plot(dcurves.k, dcurves.w_mrg.sel(h_e=i), color='gray')
            ax.plot(dcurves.k, dcurves.w_ig_nw.isel(h_e=i), color='gray')
    return ax
```

```python
fig = plt.figure(figsize=(16,16))

ci=1e-6 ; cmax=1e-5 ; cmin=-1e-5
component='symmetric'

#---non-linear contour interval for u'w'
cvals=[-0.0005,-0.00025,-0.0001,-0.00005,-0.000025,-0.00001,-0.000005,-0.0000025,
                0.0000025,0.000005,0.00001,0.000025,0.00005,0.0001,0.00025,0.0005]
cvals_cbar=[ i*10000 for i in cvals ]
ticks=[-5.0,-2.5,-1.0,-0.5,-0.25,-0.1,-0.05,-0.025,0.025,0.05,0.1,0.25,0.5,1.0,2.5,5.0]

xlim=14
ax = myplots.plotWK(fig, era5_50.cospec_uw.sel(component=component).mean('icomp'), era5_50.k, era5_50.w, ci, cmin, cmax,
                    "(a) ERA5 ($u'\omega'$)",0,0.11,0.8,0.97,contourlinescale=2, cmap='blue2red', xlim=[-1*xlim, xlim],
                    speclevs=cvals)
ax = oplot_dcurves(ax, dcurves, component=component)


ax = myplots.plotWK(fig, jra55_50.cospec_uw.sel(component=component).mean('icomp'), jra55_50.k, jra55_50.w, ci, cmin, cmax,
                    "(b) JRA55 ($u'\omega'$)",0.17,0.28,0.8,0.97,contourlinescale=2, cmap='blue2red', xlim=[-1.*xlim, xlim],
                    speclevs=cvals)
ax = oplot_dcurves(ax, dcurves, component=component)

ax = myplots.plotWK(fig, merra2_50.cospec_uw.sel(component=component).mean('icomp'), merra2_50.k, merra2_50.w, ci, cmin, cmax,
                    "(c) MERRA2 ($u'\omega'$)",0.34,0.45,0.8,0.97, contourlinescale=2, cmap='blue2red', xlim=[-1.*xlim, xlim],
                    speclevs=cvals)
ax = oplot_dcurves(ax, dcurves, component=component)

ax = cbars.plotcolorbar_sayc(fig, cvals_cbar,"$u'\omega$' ($\\times$10$^{-4}$ m Pa s$^{-1}$)",0.47,0.48,0.8,0.97,orient='vertical', ticks=ticks,
                            contourlines=True)



ci =0.2 ; cmin=-8.6 ; cmax=-7

ax = myplots.plotWK(fig, np.log10(era5_50.cospec_ww.sel(component=component).mean('icomp')), era5_50.k, era5_50.w, ci, cmin, cmax,
                    "(d) ERA5 ($\omega'\omega'$)",0,0.11,0.57,0.74,contourlinescale=1, cmap='blue2red', xlim=[-1.*xlim, xlim],
                    posonly=True)
ax = oplot_dcurves(ax, dcurves, component=component)

ax = myplots.plotWK(fig, np.log10(jra55_50.cospec_ww.sel(component=component).mean('icomp')), jra55_50.k, jra55_50.w, ci, cmin, cmax,
                    "(e) JRA55 ($\omega'\omega'$)",0.17,0.28,0.57,0.74,contourlinescale=1, cmap='blue2red', xlim=[-1.*xlim, xlim], posonly=True)
ax = oplot_dcurves(ax, dcurves, component=component)

ax = myplots.plotWK(fig, np.log10(merra2_50.cospec_ww.sel(component=component).mean('icomp')), merra2_50.k, merra2_50.w, ci, cmin, cmax,
                    "(f) MERRA2 ($\omega'\omega'$)",0.34,0.45,0.57,0.74,contourlinescale=1, cmap='blue2red', xlim=[-1.*xlim, xlim], posonly=True)
ax = oplot_dcurves(ax, dcurves, component=component)

ax = cbars.plotcolorbar_log10(fig, ci, cmin, cmax,"$\omega$'$\omega$' (Pa$^{2}$s$^{-2}$)", 0.47,0.48,0.57,0.74,orient='vertical', contourlines=True,
                              posonly=True)

fig.savefig(plotpath+'compare_reanalyses_qbocomposite_cospec.pdf', bbox_inches='tight', facecolor='white')
```
