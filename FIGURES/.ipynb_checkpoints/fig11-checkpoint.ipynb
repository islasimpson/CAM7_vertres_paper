```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from math import nan

from CASutils import seasonalplot_utils as seasplot
from CASutils import colorbar_utils as cbars
from CASutils import readdata_utils as read
from CASutils import averaging_utils as avg
```

```python
plotpath="/project/cas/islas/python_plots/CAM7_vertres_paper/FIGURES/"
```

```python
basepath="/project/cas/islas/python_savs/CAM7_vertres_paper/DATA_SORT/tape_recorder/"
dz1000 = xr.open_dataset(basepath+"dz1000_tropical_q.nc")
dz900 = xr.open_dataset(basepath+"dz900_tropical_q.nc")
dz800 = xr.open_dataset(basepath+"dz800_tropical_q.nc")
dz700 = xr.open_dataset(basepath+"dz700_tropical_q.nc")
dz600 = xr.open_dataset(basepath+"dz600_tropical_q.nc")
dz500 = xr.open_dataset(basepath+"dz500_tropical_q.nc")
dz400 = xr.open_dataset(basepath+"dz400_tropical_q.nc")
swoosh = xr.open_dataset(basepath+"SWOOSH_tropical_q_2005_2021.nc")
```

```python
fig = plt.figure(figsize=(16,16))

ci=1e-7 ; cmin=1.4e-6 ; cmax=3e-6
ax = seasplot.plot_pre_mon(fig, swoosh.Q, swoosh.pre, ci,cmin,cmax,'SWOOSH',0,0.11,0.57,0.7,taxis='month', paxis='pre', cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz1000.Q, dz1000.lev, ci,cmin,cmax,'dz1000',0.13,0.24,0.57,0.7, taxis='month',ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz900.Q, dz900.lev, ci,cmin,cmax,'dz900',0.25,0.36,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz800.Q, dz800.lev, ci,cmin,cmax,'dz800',0.37,0.48,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz700.Q, dz700.lev, ci,cmin,cmax,'dz700',0.49,0.6,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz600.Q, dz600.lev, ci,cmin,cmax,'dz600',0.61,0.72,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz500.Q, dz500.lev, ci,cmin,cmax,'dz500',0.73,0.84,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')
ax = seasplot.plot_pre_mon(fig, dz400.Q, dz400.lev, ci,cmin,cmax,'dz400',0.85,0.96,0.57,0.7, taxis='month', ylabel=False, cmap='precip_nowhite')

ax = cbars.plotcolorbar(fig, ci, cmin, cmax,'Water vapor (kg/kg)', 0.18,0.76,0.53,0.54,orient='horizontal', cmap='precip_nowhite')

fig.savefig(plotpath+'fig11.pdf', bbox_inches='tight', facecolor='white')
```
