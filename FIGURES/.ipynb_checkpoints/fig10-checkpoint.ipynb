### Plot the standard deviation of the MJO filtered vertical velocity

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from math import nan
from CASutils import mapplot_utils as mymaps
from CASutils import colorbar_utils as cbars
import sys
```

```python
plotpath="/project/cas/islas/python_plots/L83_paper/FIGURES/"
```

```python
basepath="/project/cas/islas/python_savs/L83_paper/DATA_SORT/MJOfilter/"
era5 = xr.open_dataset(basepath+'W_MJO_ERA5.nc')
dz1000 = xr.open_dataset(basepath+'W_mjo_140km_dz1000.nc')
dz900 = xr.open_dataset(basepath+'W_mjo_140km_dz900.nc')
dz800 = xr.open_dataset(basepath+'W_mjo_140km_dz800.nc')
dz700 = xr.open_dataset(basepath+'W_mjo_140km_dz700.nc')
dz600 = xr.open_dataset(basepath+'W_mjo_140km_dz600.nc')
dz500 = xr.open_dataset(basepath+'W_mjo_140km_dz500.nc')
dz400 = xr.open_dataset(basepath+'W_mjo_140km_dz400.nc')
```

```python
era5_std = era5.std('time').mean('year')
dz1000_std = dz1000.std('time').mean('year')
dz900_std = dz900.std('time').mean('year')
dz800_std = dz800.std('time').mean('year')
dz700_std = dz700.std('time').mean('year')
dz600_std = dz600.std('time').mean('year')
dz500_std = dz500.std('time').mean('year')
dz400_std = dz400.std('time').mean('year')
```

```python
fig = plt.figure(figsize=(16,16))

ci=0.004 ; cmin=-0.044 ; cmax=0.044

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, era5_std.MJO_W, era5_std.lon, era5_std.lat,ci,cmin,cmax,'ERA5',
                                                   0.0,0.11,0.85,0.94, contourlines=True)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz1000_std.MJO_W, dz1000_std.lon, dz1000_std.lat,ci,cmin,cmax,'dz1000',
                                                   0.13,0.24,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz900_std.MJO_W, dz900_std.lon, dz900_std.lat,ci,cmin,cmax,'dz900',
                                                   0.25,0.36,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz800_std.MJO_W, dz800_std.lon, dz800_std.lat,ci,cmin,cmax,'dz800',
                                                   0.37,0.48,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz700_std.MJO_W, dz700_std.lon, dz700_std.lat,ci,cmin,cmax,'dz700',
                                                   0.49,0.6,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz600_std.MJO_W, dz600_std.lon, dz600_std.lat,ci,cmin,cmax,'dz600',
                                                   0.61,0.72,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz500_std.MJO_W, dz500_std.lon, dz500_std.lat,ci,cmin,cmax,'dz500',
                                                   0.73,0.84,0.85,0.94, contourlines=True,ylabels=False)

ax = mymaps.contourmap_bothoceans_mjoregion_fill_pos(fig, dz400_std.MJO_W, dz400_std.lon, dz400_std.lat,ci,cmin,cmax,'dz400',
                                                   0.85,0.96,0.85,0.94, contourlines=True,ylabels=False)

ax = cbars.plotcolorbar(fig, ci, cmin, cmax, "$\sigma(\omega_{MJO})$ (Pas$^{-1}$)",0.18,0.78,0.82,0.83,orient='horizontal', contourlines=True,
                       posneg='pos',ticks=[0,0.008,0.016,0.024,0.032,0.04])

fig.savefig(plotpath+'mjo_140km.pdf', bbox_inches='tight', facecolor='white')
```
